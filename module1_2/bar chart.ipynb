# Bar charts


Matplotlib tem suporte para vários tipos de gráficos de barras. No caso mais geral, plotamos um gráfico de barras enviando um parâmetro dos componentes x e um parâmetro da altura da barra. Vamos fazer isso com alguns dados lineares.

```python
import matplotlib.pyplot as plt
import numpy as np

linear_data = np.array([1,2,3,4,5,6,7,8])
xvals = range(len(linear_data))

plt.bar(xvals,linear_data,width=0.3)
```

Para adicionar uma segunda barra, nós simplesmente chamamos `plt.bar` novamente, lembrando que precisamos ajustar o componente x para compensar a primeira barra que plotamos.

```python
new_xvals = []
exponential_data = linear_data**2

for item in xvals:
    new_xvals.append(item+0.3)

plt.bar(xvals, linear_data, width = 0.3)
plt.bar(new_xvals, exponential_data, width = 0.3 ,color='red')
```

Portanto, o plot parece bom, mas não ótimo. Todos os rótulos x estão à esquerda dos itens que estão sendo descritos e não centralizados. Agora, podemos centralizá-los com uma determinada barra, usando o parâmetro `align`, mas é um pouco frustrante quando você tem que fazer várias barras, e você mesmo tem que calcular. E é difícil adicionar uma nova série de dados manualmente. E isso piora, na verdade temos que iterar manualmente e adicionar datas se quisermos. Porque, caso contrário, não poderíamos definir o local esquerdo da barra adequadamente.

Existem alguns outros recursos interessantes de gráficos de barras. Por exemplo, você também pode adicionar barras de erro a cada barra, usando o parâmetro y-error. Por exemplo, cada um de nossos dados nos dados lineares pode, na verdade, ser um valor médio, calculado a partir de muitas observações diferentes. Então, poderíamos apenas criar uma lista de valores de erro que queremos renderizar também e mostrá-los.

```python
from random import randint
linear_err = [randint(1,4) for x in range(len(linear_data))] 

plt.bar(xvals, linear_data, width = 0.3, yerr=linear_err)
```

Também podemos fazer gráficos de barras empilhadas. Por exemplo, se quiséssemos mostrar valores cumulativos e ao mesmo tempo manter a série independente, poderíamos fazer isso definindo o parâmetro bottom e nosso segundo gráfico para ser igual ao primeiro conjunto de dados a plotar.

```python
xvals = range(len(linear_data))
plt.bar(xvals, linear_data, width = 0.3, color='b')
plt.bar(xvals, exponential_data, width = 0.3, bottom=linear_data, color='r')
```

Por fim, podemos transformar esse gráfico de barras em um gráfico de barras horizontal chamando a função barh.

Mas observe que temos que mudar o fundo para a esquerda e a largura para a altura.

```python
plt.barh(xvals, linear_data, height = 0.3, color='b')
plt.barh(xvals, exponential_data, height = 0.3, left=linear_data, color='r')
```
