Um scatter plot é um plot bidimensional similar ao plot de linha. A função scatter recebe um valor x-axis como primeiro argumento e um valor y-axis como segundo.

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

# Vamos criar alguns dados falsos
x=np.array([1,2,3,4,5,6,7,8])
y=x

# Agora vamos criar um scatter plot. Isso é similar ao plt.plot(),
# mas os objetos filhos subjacentes nos eixos não são Line2D.
teste=plt.scatter(x,y)
```

```python
# Podemos passar uma lista de cores para cada ponto.
# Vamos fazer todos verdes, menos o último

colors = ['green']*(len(x)-1)
colors.append('red')
plt.scatter(x,y,s=100,c=colors)
```

```python
# Agora, vamos pegar as duas listas e plotar separadamente.
# Ao invés de plotar como uma série de dados, vamos fatiar as listas e 
# plotar como duas séries de dados

# E também , ao invés de escolher uma cor por dado, vamos setar uma cor por série.

plt.scatter(x[:2],y[:2],s=100,c='red',label='Tall students')
plt.scatter(x[2:],y[2:],s=100,c='blue',label='Short students')
```

```python
# Labels, titles and legend

plt.scatter(x[:2],y[:2],s=100,c='red',label='Tall students')
plt.scatter(x[2:],y[2:],s=100,c='blue',label='Short students')

plt.xlabel('The number of times the child kicked a ball')
plt.ylabel('The grade of the student')
plt.title('Relationship between ball kicking and grades')
 
plt.legend()
```

 Na documentação de `legend` vamos ver que existe um número de parâmetros que podem ser passados. Um deles é o parâmetro loc. Passando loc=4, estaremos colocando a legenda na parte inferior direita.
 
Podemos, também, nos livrar do quadro(frameon) e adicionar um título a legenda.

```python
f=plt.figure()
plt.scatter(x[:2],y[:2],s=100,c='red',label='Tall students')
plt.scatter(x[2:],y[2:],s=100,c='blue',label='Short students')

plt.xlabel('The number of times the child kicked a ball')
plt.ylabel('The grade of the student')
plt.title('Relationship between ball kicking and grades')
 
plt.legend(loc=4,frameon=False,title='Legend')
```

Todas essas coisas são objetos renderizáveis associado com os eixos. A legenda, por exemplo, é um artista, o que significa que possui filhos. Vamos criar uma função que retorna os filhos de um artista.

```python
# Primeiro importamos a classe 'artista' do matplotlib
from matplotlib.artist import Artist

# Então, vamos fazer uma função recursiva que recebe o artista e um parâmetro de profundidade.
def rec_gc(art,depth=0):
    # Checa se o objeto é um artista
    if isinstance(art,Artist):
        # Aumenta a profundidade para melhor visualização
        print("  " * depth + str(art))
        for child in art.get_children():
            rec_gc(child, depth+2)


rec_gc(f.legend())
```
