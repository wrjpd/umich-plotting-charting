# Line Plots

Um line plot é criado, também, com a função `plot` e renderiza uma série de pontos que são conectados por uma linha.

**line plots x scatter plots:** 

- Primeiro, nós podemos passar somente o y-axes, não o x-axes. Assim, a função plot reconhece que queremos usar o indice como eixo x.

- Em segundo lugar, ao passarmos x e y, vemos que o gráfico identifica isso como duas séries de dados e que as cores dos dados da série são diferentes, incluindo os pontos de dados e as linhas entre os pontos de dados. Isso é diferente do gráfico de dispersão, que exige que rotulemos as linhas diretamente.



```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

linear_data = np.array([1,2,3,4,5,6,7,8])
exponential_data = linear_data**2
```

```python
# Podemos plotar somente umas das séries
plt.plot(linear_data)
```

```python
# Podemos fazer exponential data x linaer data

plt.plot(linear_data,exponential_data)
```

```python
# Podemos fazer dois plots diferentes na mesma figure

plt.plot(linear_data,'-o',exponential_data,'-o')

# Podemos adicionar um terceiro
plt.plot([22,45,55])
```

```python
# Podemos usar as funções regulares de eixo para criar 
# labels e títulos

plt.xlabel('Some data')
plt.ylabel('Some other data')
plt.title('A title')
plt.legend(['Baseline', 'Competition', 'Us'])

plt.plot(linear_data, '-o', exponential_data, '-o')
plt.plot([22,44,55], '--r')
plt.legend()
```


### Fill between

Essa função não é específica para line plots, mas é mais utilizada com line plots. Vamos imaginar que queremos explicitar a diferença entre duas curvas, podemos pintar uma cor entre essas curvas usando a função fillbetween

```python
plt.xlabel('Some data')
plt.ylabel('Some other data')
plt.title('A title')
plt.legend(['Baseline', 'Competition', 'Us'])
plt.plot(linear_data, '-o', exponential_data, '-o')

# Precisamos pegar o objeto do atual Axes e chamar a função fillbetween.
# Não passamos nenhum x-axis, então vamos usar o mesmo range dos dados.
# Vamos passar também o limite inferior e superior.

print(plt.gca()) #AxesSubplot(0.125,0.11;0.775x0.77)

plt.gca().fill_between(range(len(linear_data)), 
                       linear_data, exponential_data, 
                       facecolor='blue', 
                       alpha=0.25)
```

## Datas

As datas se sobrepõem fortemente. Agora, há algumas coisas que poderíamos fazer aqui. Por exemplo, tudo é de 2017. Portanto, poderíamos iterar e alterar os rótulos removendo o ano e, em seguida, adicionar um rótulo de eixo x que esclarece que é tudo para o ano de 2017.

Existem muitas propriedades interessantes do objeto axes e você deve usar algumas delas na atribuição. Por exemplo; você pode obter as linhas de grade, os locais dos ticks principais e secundários e assim por diante. Assim como todos os artistas, um axes tem um monte de filhos que também são artistas.

Podemos acessar o texto dos ticks usando a função get tick labels. Cada um dos rótulos de marcação é um objeto de texto que é um artista. Isso significa que você pode usar várias funções de artista diferentes e uma específica para o texto é a função de rotação definida, que altera a rotação com base em graus. Vamos percorrer os rótulos dos eixos e mudar isso.

```python
help(plt.gca)
```

```python
# First, we'll import the artist class from matplotlib.
from matplotlib.artist import Artist

# Then we'll make a recursive function which takes in an artist and some depth parameter.
def rec_gc(art, depth=0):
    print(art)    
    
    # Then checks if the object is an artist and if so, prints out its string name.
    if isinstance(art, Artist):
        # increase the depth for pretty printing
        print("  " * depth + str(art))
        for child in art.get_children():
            rec_gc(child, depth+2)

# Call this function on the legend artist to see what the legend is made up of
rec_gc(plt.gca())
```

```python
# Vamos acessar os eixos

observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
plt.plot(observation_dates, linear_data, '-o',  observation_dates, exponential_data, '-o');

x=plt.gca().xaxis
for item in x.get_ticklabels():
    print(type(item))
```

```python
# plt.gca é uma função do módulo matplotlib.pyplot
# plt.gca() retorna uma classe
# plt.gca() possui uma classe xaxis
# xaxis possui o método get_ticklabels que é um método da xaxis
# get_ticklabels retorna uma lista contendo instância da classe matplotlib.text.Text
# matplotlib.text.Text possui a função set_rotation
```

```python
# Podemos pegar o eixo e rotacionar.
observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
plt.plot(observation_dates, linear_data, '-o',  observation_dates, exponential_data, '-o');

x = plt.gca().xaxis

for item in x.get_ticklabels():
    item.set_rotation(45)
    
plt.xlabel('oi')
```

### Latex e tamanho

```python
plt.figure(figsize=(8,6))


observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
plt.plot(observation_dates, linear_data, '-o',  observation_dates, exponential_data, '-o');
x = plt.gca().xaxis
# rotate the tick labels for the x axis
for item in x.get_ticklabels():
    item.set_rotation(45)
ax = plt.gca()
ax.set_xlabel('Date')
ax.set_ylabel('Units')

# And finally, I'll add in some mathematical expressions which you can do 
# in any text element
ax.set_title("Exponential ($x^2$) vs. Linear ($x$) performance");
```
